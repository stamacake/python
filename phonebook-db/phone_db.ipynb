```python
#phonelist = [['Anton', 'Apple', '1234567','MTC','115 rub','no bank acc '],
#            ['Denis', 'Samsung', '22234567','Megafon','445 rub','from mtc'],
#            ['Alex', 'Xiaomi', '123555577','Beeline','15 usd','rich']]

from tkinter import *

import pandas as pd
#df = pd.DataFrame(phonelist)
#df.to_csv('data11.csv',index=False)
df1 = pd.read_csv('data11.csv')
phonelist  =  df1.values.tolist()
print(phonelist)


def which_selected():
    print("At {0}".format(select.curselection()))
    ae = select.curselection()
    return ae[0]


def add_entry():
    phonelist.append([fnamevar.get(), lnamevar.get(), phonevar.get(),tar1.get(),pay1.get(),etc1.get()])
    set_select()
    df = pd.DataFrame(phonelist)
    df.to_csv('data11.csv',index=False)

def update_entry():
    phonelist[which_selected()] = [fnamevar.get(),lnamevar.get(),phonevar.get(),tar1.get(),pay1.get(),etc1.get()]
    df = pd.DataFrame(phonelist)
    df.to_csv('data11.csv',index=False)
    df = pd.DataFrame(phonelist)
    df.to_csv('data11.csv',index=False)

def delete_entry():
    del phonelist[which_selected()]
    set_select()
    df = pd.DataFrame(phonelist)
    df.to_csv('data11.csv',index=False)


def load_entry():
    fname, lname, phone, tar2, pay2, etc2 = phonelist[which_selected()]
    fnamevar.set(fname)
    lnamevar.set(lname)
    phonevar.set(phone)
    tar1.set(tar2)                 
    pay1.set(pay2) 
    etc1.set(etc2)                  

def make_window():
    global fnamevar, lnamevar, phonevar, tar1, pay1, etc1, select
    win = Tk()

    frame1 = Frame(win)
    frame1.pack()

    Label(frame1, text="Name").grid(row=0, column=0, sticky=W)
    fnamevar = StringVar()
    fname = Entry(frame1, textvariable=fnamevar)
    fname.grid(row=0, column=1, sticky=W)

    Label(frame1, text="Phone model").grid(row=1, column=0, sticky=W)
    lnamevar = StringVar()
    lname = Entry(frame1, textvariable=lnamevar)
    lname.grid(row=1, column=1, sticky=W)

    Label(frame1, text="Phone number").grid(row=2, column=0, sticky=W)
    phonevar = StringVar()
    phone = Entry(frame1, textvariable=phonevar)
    phone.grid(row=2, column=1, sticky=W)
    
    
    Label(frame1, text="Tarif").grid(row=3, column=0, sticky=W)
    tar1 = StringVar()
    tar2 = Entry(frame1, textvariable=tar1)
    tar2.grid(row=3, column=1, sticky=W)
    
    Label(frame1, text="Payment").grid(row=4, column=0, sticky=W)
    pay1 = StringVar()
    pay2 = Entry(frame1, textvariable=pay1)
    pay2.grid(row=4, column=1, sticky=W)
    
    Label(frame1, text="ETC").grid(row=5, column=0, sticky=W)
    etc1 = StringVar()
    etc2 = Entry(frame1, textvariable=etc1)
    etc2.grid(row=5, column=1, sticky=W)

    frame2 = Frame(win)       # Row of buttons
    frame2.pack()
    b1 = Button(frame2, text=" Add  ", command=add_entry)
    b2 = Button(frame2, text="Update", command=update_entry)
    b3 = Button(frame2, text="Delete", command=delete_entry)
    b4 = Button(frame2, text="Load  ", command=load_entry)
    b5 = Button(frame2, text="Refresh", command=set_select)
    b1.pack(side=LEFT)
    b2.pack(side=LEFT)
    b3.pack(side=LEFT)
    b4.pack(side=LEFT)
    b5.pack(side=LEFT)

    frame3 = Frame(win)       # select of names
    frame3.pack()
    scroll = Scrollbar(frame3, orient=VERTICAL)
    select = Listbox(frame3, yscrollcommand=scroll.set, height=6,width=30)
    scroll.config(command=select.yview)
    scroll.pack(side=RIGHT, fill=Y)
    select.pack(side=LEFT, fill=BOTH, expand=1)
    return win


def set_select():
    #phonelist.sort(key=lambda record: record[1])
    select.delete(0, END)
    for fname, lname, phone,tar2,pay2,etc2 in phonelist:
        select.insert(END,"{0},{1},{2},{3},{4},{5}".format(fname,lname,phone,tar2,pay2,etc2))


win = make_window()
set_select()
win.mainloop()
```
